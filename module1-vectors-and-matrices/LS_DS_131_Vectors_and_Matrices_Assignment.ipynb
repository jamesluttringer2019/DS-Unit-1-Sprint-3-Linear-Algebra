## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = np.array([[12,11],
              [7,10]])
J = np.array([[0,1,2],
              [7,10,4],
              [3,2,0]])

det_H = np.linalg.det(H)
det_J = np.linalg.det(J)

print('|H|: '+str(det_H))
print('|J|: '+str(det_J))
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
H_inv = np.linalg.inv(H)
print('H Inverse: '+str(H_inv))
J_inv = np.linalg.inv(J)
print('J Inverse: '+str(J_inv))
```
